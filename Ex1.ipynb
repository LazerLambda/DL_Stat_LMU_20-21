```python
import numpy as np
import math
import matplotlib.pyplot as plt
from mpl_toolkits.axisartist import SubplotZero
```

```python
# prepare data
x = np.array([[0,0], [1,0], [0,-1], [-1,0], [0,1]])
unpack_x = lambda x, i, slice : [e[i] for e in x[slice]] 

# Show figure 1
fig, ax = plt.subplots()
ax.plot(unpack_x(x, 0, [0,1,2]), unpack_x(x, 1, [0,1,2]), "go")
ax.plot(unpack_x(x, 0, [3,4]), unpack_x(x, 1, [3,4]), "go", c="coral")
ax.set_xlim(-2, 2)
ax.set_ylim(-2, 2)

plt.show()
```

```python
X =  np.c_[np.ones((5,1)), x]
sigmoid = lambda arg : 1 / (1 + np.exp(-arg))
```

```python
a0 = -3
a1 = -5
a2 = 5
w = np.array([a0, a1, a2])

# Classification:
sigmoid(X @ w)
```

```python
# Show figure 2
fig, ax = plt.subplots()
ax.plot(unpack_x(x, 0, [0,3,4]), unpack_x(x, 1, [0,3,4]), "go")
ax.plot(unpack_x(x, 0, [1,2]), unpack_x(x, 1, [1,2]), "go", c="coral")
ax.set_xlim(-2, 2)
ax.set_ylim(-2, 2)
```

```python
b0 = -3
b1 = 5
b2 = -5
w = np.array([b0, b1, b2])

# Classification
sigmoid(X @ w)
```

```python
# Transformation
x = np.array([[0,0], [0,1], [0,1], [1,0], [1,0]])

# Show figure 3
fig, ax = plt.subplots()
ax.plot(unpack_x(x, 0, [0,3,4]), unpack_x(x, 1, [0,3,4]), "go")
ax.plot(unpack_x(x, 0, [1,2]), unpack_x(x, 1, [1,2]), "go", c="coral")
ax.set_xlim(-0.2, 2)
ax.set_ylim(-0.2, 2)
```

```python
# transformed data
Z = np.c_[np.ones((5,1)), np.zeros((5,1)), np.array([0,1,1,1,1])]

g0 = 10
g1 = 1
g2 = -15
w = np.array([g0, g1, g2])

# Classification
sigmoid(Z @ w)
```
